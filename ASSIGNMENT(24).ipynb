```python

import re
import matplotlib.pyplot as plt
from collections import Counter

def extract_emojis_and_plot(text):
    """
    Extracts emojis from a string and creates a bar chart of their frequencies.

    Args:
        text: The input string.
    """

    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        u"\U0001F1E0-\U0001F1FF"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)
    emojis = emoji_pattern.findall(text)
    emoji_counts = Counter(emojis)
    if not emoji_counts:
        print("No emojis found in the input text.")
        return

    plt.figure(figsize=(10, 6))
    plt.bar(emoji_counts.keys(), emoji_counts.values())
    plt.xlabel("Emojis")
    plt.ylabel("Frequency")
    plt.title("Emoji Frequency")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()



text_with_emojis = "Hello! üòÑüëç This is a test string with some emojis like üòÇ and ‚úåÔ∏è.  I also like üéâ."
extract_emojis_and_plot(text_with_emojis)
```
